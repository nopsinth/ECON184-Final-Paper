Return on Equity

```python
df_roe = pd.read_csv('spy_roe.csv')
```

```python
val = df_roe.values[:, 2:].astype(np.float)
```

```python
val_cap = np.minimum(np.maximum(val, np.nanpercentile(val, 5, axis=0)), np.nanpercentile(val, 95, axis=0))
```

```python
val_z = (val_cap - np.nanmean(val_cap, axis=0)) / np.nanstd(val_cap, axis=0)
```

```python
val_z.astype(str)
```

```python
v = df_roe.values
v[:, 2:] = val_z.astype(str)
df_roe[:] = v
```

Accruals Ratio

```python
df_noa = pd.read_csv("spy_noa.csv")
df_noa.drop(df_noa.columns[len(df_noa.columns)-1], axis=1, inplace=True)
```

```python
val = df_noa.values[:, 2:].astype(np.float)
```

```python
accrual_val = []
for i in range(1, val.shape[1]):
    diff_noa = val[:,i] - val[:, i-1]
    avg_noa = (val[:, i] + val[:, i-1])/2
    accrual_val.append(diff_noa/avg_noa)

accrual_val = np.array(accrual_val)
accrual_val = accrual_val.T
```

```python
v = df_noa.values
v[:, 3:] = accrual_val.astype(str)
df_noa[:] = v
```

```python
df_noa
```

```python
val_cap = np.minimum(np.maximum(val, np.nanpercentile(val, 5, axis=0)), np.nanpercentile(val, 95, axis=0))
val_z = -(val_cap - np.nanmean(val_cap, axis=0)) / np.nanstd(val_cap, axis=0)
v = df_noa.values
v[:, 2:] = val_z.astype(str)
df_noa[:] = v
```

```python
df_noa.to_csv("mod_spy_noa.csv")
```

```python
df_roe = pd.read_csv("mod_spy_roe.csv")
df_leverage = pd.read_csv("mod_spy_dtoe.csv")
df_noa = pd.read_csv("mod_spy_noa.csv")
```

```python
df_roe.drop(df_noa.columns[0], axis=1, inplace=True)
df_leverage.drop(df_noa.columns[0], axis=1, inplace=True)
df_noa.drop(df_noa.columns[0], axis=1, inplace=True)
```

```python
df_leverage.shape
```

```python
val_z_roe = df_roe.values[:, 2:].astype(np.float)
val_z_leverage = df_leverage.values[:, 2:].astype(np.float)
val_z_noa = df_noa.values[:, 2:].astype(np.float)
```

```python
import matplotlib.pyplot as plt
```

```python
(val_z_noa.shape)
```

```python
average_val_z = (np.nan_to_num(val_z_roe) + np.nan_to_num(val_z_leverage) + np.nan_to_num(val_z_noa))/ (3 - np.isnan(val_z_roe) - np.isnan(val_z_leverage) - np.isnan(val_z_noa)) #nan_to_num
```

```python
plt.plot(val_z_leverage[0, 3:])
```

```python
val_q = (1/(1-average_val_z)*(average_val_z < 0) + (1 + average_val_z)*(average_val_z > 0))
```

```python
df_q = df_roe
```

```python
val_q.astype(str)
v = df_q.values
v[:, 2:] = val_q.astype(str)
df_q[:] = v
df_q.to_csv("df_q.csv")
```

```python
val_q_leverage = (1/(1-val_z_leverage)*(val_z_leverage < 0) + (1 + val_z_leverage)*(val_z_leverage > 0))
val_q_roe = (1/(1-val_z_roe)*(val_z_roe < 0) + (1 + val_z_roe)*(val_z_roe > 0))
val_q_noa = (1/(1-val_z_noa)*(val_z_noa < 0) + (1 + val_z_noa)*(val_z_noa > 0))
```

```python
val_q_noa.shape
```

```python
df_q_leverage = df_roe
val_q_leverage.astype(str)
v = df_q_leverage.values
v[:, 2:] = val_q_leverage.astype(str)
df_q_leverage[:] = v
df_q_leverage.to_csv("df_q_leverage.csv")
```

Cumulative Return (Total Return Index)

```python
data_return = pd.read_csv("spy_total_return_indx.csv")
```

```python
data_return.columns[21:33]
```

```python
val = data_return.values[:, 21:33].astype(np.float) #2000 - 2002
```

```python
for i in range(1, 12):
    cum_return = (val[:,i] - val[:, 0])/(val[:, 0])
    val[:, i] = cum_return
    
val[:, 0] = 0
```

```python
v = data_return.values
v[:, 21:33] = val.astype(str)
data_return[:] = v
```

```python
data_return_02 = data_return[data_return.columns[[0] + list(range(21, 33, 1))]]
```
